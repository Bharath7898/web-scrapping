```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
url = "https://en.wikipedia.org/wiki/List_of_fatalities_while_playing_cricket"   # open link
html = urlopen(url)
```

```python
soup = BeautifulSoup(html)
```

```python
title = soup.title
print(title)
print(title.text)
```

```python
links = soup.find_all('a', href = True)
for link in links:
    print(link['href'])
```

```python
allrows = soup.find_all('tr')
print(allrows)
```

```python
for row in allrows:
    row_list = row.find_all('td')
print(row_list)
```

```python

for cell in row_list:
    print(cell.text)
```

```python
data = []
allrows = soup.find_all('tr')
for row in allrows:
    row_list = row.find_all('td')
    dataRow = []
    for cell in row_list:
        dataRow.append(cell.text)
    data.append(dataRow)
print(data)
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
col_headers = soup.find_all('th')
col_headers
```

```python
header_list = []
col_headers = soup.find_all('th')
for col in col_headers:
    header_list.append(col.text)
print(header_list)
```

```python
df.columns = header_list
print(df.head)
```
